## Data Acquisition and Preprocessing

### Data Sources

1. Communities and Crime Dataset
The Communities and Crime Unnormalized dataset was sourced from the UCI Machine Learning Repository (https://archive.ics.uci.edu/dataset/211/communities+and+crime+unnormalized). This dataset combines socio-economic data from the 1990 Census, law enforcement data from the 1990 Law Enforcement Management and Administrative Statistics (LEMAS) survey, and crime data from the 1995 FBI Uniform Crime Reporting (UCR) program. It includes 2,215 instances and 125 features, covering various attributes such as population demographics, income levels, and law enforcement presenceThe data was originally in a TEXT format and required conversion to CSV format to facilitate easier manipulation and analysis using Python’s Pandas library.
<br>

2. Criminal Victimization Dataset
The Criminal Victimization in the 22 Largest U.S. States dataset (https://bjs.ojp.gov/library/publications/criminal-victimization-22-largest-us-states-2017-2019) was accessed from the Bureau of Justice Statistics (BJS). This dataset presents state-level estimates of violent and property victimization for the aggregate years of 2017 to 2019. It captures the experiences of individuals and households, offering a more nuanced view of crime that includes both reported and unreported incidents. The data was provided in separate files for different features, which required combining these files into a single cohesive dataset for analysis.
<br>

3. US Cities Database
The US Cities dataset was obtained from SimpleMaps (https://simplemaps.com/data/us-cities), which provides a simple and accurate database of U.S. cities. This dataset is extensive and up-to-date, containing information about over 109,000 cities and towns across the United States. It includes vital fields such as city names, state identifiers, population estimates, and various socio-economic indicators. The data was readily available in CSV format, making it straightforward to import into our analysis environment.
<br>

Since there are a large number of variables in the overall data, we will provide detailed variable descriptions of the selected features in subsequent chapters.

### Data Preparation Steps

#### Acquiring and Structuring Data

The initial data cleaning for the Communities and Crime dataset involved removing extraneous characters from the nonViolPerPop and communityname columns. Because the communityname entries were formatted as Nametype, the community type (i.e., town, city, borough, township) was separated by space, and a new column titled communityName was created with just the name of the community. Lastly, a space was added before capital letters in the communityName. For example, the community “WestSacramento” was transformed to “West Sacramento.”

<br>
The Criminal Victimization dataset required concatenating several CSV files into one DataFrame. After combining these files, because one of the columns included whether the observation was significant according to a 95% and 90% confidence interval compared to the values across the US, this column was expanded into two boolean columns “95%CIsig” and “90%CIsig.” These columns have a TRUE value if the data is significant at the respective confidence level. Then, we adjusted the column types for consistency and reordered them for ease of use. The prefix “state” was added to each column to specify that this was statewide data and not applicable to the individual communities in the Communities and Crime dataset. Lastly, we renamed the columns to reflect the specific type of criminal victimization being measured, ensuring clarity in our merged dataset.


<br>

For the US Cities dataset, minimal cleaning was performed since it was relatively clean from the outset. We focused on dropping unnecessary columns that did not contribute to our analysis.

#### Merging Datasets

With the three datasets cleaned and structured, we proceeded to merge them. The Communities and Crime dataset was merged with the US Cities database using the community name and state identifiers as keys. This merged dataset was then further merged with the top 22 state dataframe on state_name and state. Redundant columns were dropped, and the final merged data was exported as a CSV for ease of dispersal. To account for potential naming conflicts due to cities with similar names across different states, we ensured that the merges were conducted carefully to avoid data misalignment.

Before diving into cleaning, we began by inspecting the structure of the datasets. Using the Pandas info() method, we reviewed the number of entries, column names, and data types. This initial inspection helped us identify potential issues such as missing values and inconsistencies within the datasets.

#### Handling Missing Values

To address missing values in the dataset, we selected a single method to ensure consistency and avoid redundancy. After evaluating different approaches, we decided to use K-Nearest Neighbors (KNN) imputation as it leverages the relationships between similar data points to estimate missing values more effectively. This method was particularly useful for handling both numerical and categorical variables, ensuring that missing values were inferred based on the closest data observations.

To justify our selection, we examined the distribution of two key numerical columns, Median Income (medIncome) and Percentage of Population in Poverty (pctPoverty), by plotting their histograms. If the distribution was symmetrical, mean imputation would have been a viable option. 

```python
from IPython.display import Image, display
display(Image("Median Income (medIncome) Histogram.png"))
```

```python
from IPython.display import Image, display
display(Image("Percentage of Population in Poverty (pctPoverty).png"))
```

Upon analysis, we found that these distributions were skewed, making KNN imputation the better choice. This approach ensured that the estimated values preserved the underlying data structure and minimized bias in the analysis.

Through this method, we demonstrated a clear understanding of missing data treatment and ensured that imputation methods were appropriately applied based on the data characteristics.

#### Removing Duplicates

After handling missing values, we checked for duplicate rows within the datasets. Duplicates can skew results and lead to misleading conclusions. We utilized the duplicated() method to identify duplicate entries and then removed them using the drop_duplicates() method. This step ensured that each entry in our dataset was unique and representative of the data we aimed to analyze.

#### Data Transformation

Data transformation was another key aspect of our preprocessing steps. This involved several actions:
<br>

- Column Renaming: To improve clarity, columns were renamed to reflect their content more accurately. For instance, columns related to crime rates received prefixes indicating whether they pertained to state or community data.
- Data Type Adjustments: Each column was assigned an appropriate data type. Categorical variables were converted to the category type, while numerical columns were set to float or int as needed. This adjustment is critical for performing accurate analyses and visualizations later on.
- String Formatting: For columns with community names, extraneous characters were removed, and consistent formatting was ensured. For example, community names formatted as “NameType” were split to create a clean column with just the community name, enhancing readability.

In our analysis, we focused on identifying and handling outliers in key numerical columns, using median income (medIncome) and percentage of poverty (pctPoverty) as examples.

```python
from IPython.display import Image, display
display(Image("Median Income (medIncome) Boxplot.png"))
display(Image("Percentage of Population in Poverty (pctPoverty) Boxplot.png"))
```

We began by visualizing the distributions of medIncome and pctPoverty using boxplots. Boxplots are effective for highlighting the presence of outliers by displaying the interquartile range (IQR) and any points that fall outside the whiskers. To manage outliers, we applied a winsorization technique, capping extreme values at the 99th percentile. This approach allowed us to mitigate the impact of outliers without entirely discarding valuable data points.

Before analysis, significant data preparation steps were necessary to ensure data quality and consistency. The Communities and Crime dataset required conversion from text format to CSV and included the removal of extraneous characters in key columns. The Criminal Victimization dataset was fragmented across multiple files, requiring concatenation and restructuring. To distinguish statewide statistics from community-level data, state-specific columns were prefixed accordingly. The US Cities dataset required minimal cleaning, as it was well-structured from the outset.

Once cleaned, datasets were merged based on common identifiers such as community names and state codes. Given that cities with the same name exist in different states, careful validation was performed to prevent data misalignment. Handling missing values was crucial, with numerical attributes imputed using median values and categorical variables filled using mode-based imputation. Additionally, K-Nearest Neighbors (KNN) imputation was applied to enhance the estimation of missing values based on neighboring observations. Outlier detection techniques were also implemented to identify and manage extreme values in key variables such as median income and crime rates.

## Feature Engineering and Dimensionality Reduction

#### Standardization

Standardization is a crucial step in preprocessing numerical features, ensuring that all variables are scaled to have a mean of 0 and a standard deviation of 1. This transformation eliminates potential biases due to differing magnitudes among features, making them more suitable for machine learning models.

#### Label Encoding

Categorical variables were converted into numerical representations using label encoding. This transformation ensures that algorithms, particularly those that require numerical input, can effectively process categorical attributes. We chose this method instead of one-hot encoding because the categorical columns are either ordinate or have some similar categories (e.g. State) so label encoding would make more sense.

#### Dimensionality Reduction with PCA

Variance proportion explained by each principal component.：

- PC1: 0.2052
- PC2: 0.1386
- PC3: 0.0862
- PC4: 0.0759
- PC5: 0.0520

To address high-dimensionality issues in the dataset, Principal Component Analysis (PCA) was employed. This technique helps retain the most significant variance in the data while reducing redundant features, thereby improving computational efficiency and reducing the risk of  overfitting. 

The results indicate that the first two principal components alone account for over 35% of the total variance, which significantly simplifies subsequent modeling efforts.

The scatter plot below illustrates the distribution of data in the reduced-dimensional space, showing how the first two principal components capture the most critical variance patterns.

## Key Findings

#### Economic and Crime Relationship

A moderate positive correlation (0.25) was observed between the poverty rate (pctPoverty) and the murder rate, indicating that communities with higher poverty levels may experience increased violent crime. This relationship highlights the potential influence of socio-economic challenges on crime prevalence. The correlation matrix further reinforces this connection, demonstrating how economic deprivation can contribute to higher crime rates.

Additionally, a weak negative correlation (-0.17) exists between median income (medIncome) and the murder rate. This suggests that communities with higher median incomes tend to experience lower murder rates, aligning with existing research on the protective effects of economic stability. The correlation matrix confirms that various economic indicators influence crime rates, further emphasizing the need for targeted economic interventions to reduce crime.

#### Demographics and Crime Relationship

The correlation matrix reveals a positive correlation (0.20) between population density (popDensity) and the murder rate. This suggests that densely populated areas may experience higher crime rates due to factors such as increased anonymity, social tensions, and economic disparity. Urban environments often provide greater opportunities for criminal activity, which can contribute to elevated crime rates. Understanding these dynamics allows policymakers to implement crime prevention strategies tailored to densely populated communities.

#### Public Safety and Crime Relationship

Law enforcement presence also plays a significant role in crime rates. The number of police officers (numPolice) and police budget per capita (policBudgetPerPop) exhibit moderate correlations with crime rates in the correlation matrix. Communities with a greater police presence generally report lower crime rates, reinforcing the importance of law enforcement in crime deterrence. This dimensionality reduction allows for a more focused analysis of the most influential factors affecting crime rates while minimizing noise from less significant features.

## Challenges Faced

During the course of this project, several challenges emerged in data acquisition, processing, and analysis. One of the primary difficulties was handling data inconsistencies across different sources. The Communities and Crime dataset required extensive cleaning and restructuring to align with the format of the other datasets. Additionally, merging datasets proved to be complex due to variations in naming conventions and differences in the granularity of geographical identifiers, requiring careful validation to ensure accurate data integration.

Another challenge involved missing data, particularly in key socio-economic indicators and crime statistics. While K-Nearest Neighbors (KNN) imputation was employed to estimate missing values, ensuring the integrity of these estimations remained a priority. Outliers in economic and crime-related variables also needed to be addressed using winsorization to prevent skewed analyses. Computational efficiency was another hurdle, particularly in performing Principal Component Analysis (PCA) and correlation analysis on a large dataset. Optimizing these processes required significant resource allocation and careful selection of methods to balance accuracy and efficiency.
