## Handle Missing Values

Missing values can be handled using median/mode or KNN method. After handling missing values, 

```python
# Handling missing values using median 
df_med = df.copy(deep=True)
missing_counts = df_med.isnull().sum()
print(missing_counts)
num_cols = df_med.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df_med.select_dtypes(include=['object']).columns.tolist()
df_med[num_cols] = df_med[num_cols].fillna(df[num_cols].median())
for col in cat_cols:
    df_med[col] = df_med[col].fillna(df[col].mode()[0])
df_med.describe().to_csv("descriptive_stats.csv")
```

```python
# Handling missing values using KNN 
from sklearn.impute import KNNImputer
df_knn = df.copy(deep=True)
# num_cols = df_knn.select_dtypes(include=[np.number]).columns #redundant line

# Initialize the imputer
imputer = KNNImputer(n_neighbors=5)
df_knn[num_cols] = imputer.fit_transform(df_knn[num_cols])

# Imputation for categorical columns 
# cat_cols = df_knn.select_dtypes(include=["object"]).columns #redundant line
for col in cat_cols:
    if df_knn[col].isnull().sum() > 0:
        df_knn[col] = df_knn[col].fillna(df_knn[col].mode()[0])
```

## Feature Transformation

```python
from scipy.stats import boxcox
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Numerical Data Transformations
num_cols = df.select_dtypes(include=[np.number]).columns

# For Box-Cox transformation, the data must be positive.
for col in ['population','medIncome']:
    if col in df.columns:
        # Check for non-positive values and shift if necessary
        min_val = df[col].min()
        if min_val <= 0:
            df[col] = df[col] + abs(min_val) + 1
        transformed, lam = boxcox(df[col])
        df[col + '_boxcox'] = transformed
        print(f"Applied Box-Cox transformation on {col} (lambda:{lam:.4f}).")
        # Optionally, drop or retain the original column
        
        
# Standardize numerical features
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])


# Normalize (MinMax scaling)
minmax = MinMaxScaler()
df[num_cols] = minmax.fit_transform(df[num_cols])
```
